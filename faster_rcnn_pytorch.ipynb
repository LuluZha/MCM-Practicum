#Delete ImageSets/Main/.txt files

```
import os#首先导入os模块，进行文件的查找，修改，删除等操作一般都要事先导入os模块

def del_files(path):
  for root,dirs,files in os.walk(path):
    for name in files:
      if '.txt' in name:
        os.remove(os.path.join(root,name))##os.move语句为删除文件语句

      print('Delete files:',os.path.join(root,name))   
if __name__=='__main__':
  path=r'/content/drive/My Drive/faster-rcnn.pytorch/data/VOCdevkit2007/VOC2007/ImageSets/Main'#此为需要删除的路径
  del_files(path)#调用函数
```
